```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ecomerce-data.csv')
df
```

```python
'''
1.) จงหาลูกค้าที่ซื้อสินค้าบ่อยที่สุด 5 คนแรก (คนที่มีหลาย invoice)
hint: groupby, sort_values, nunique, head
'''

df.groupby('CustomerID')[['InvoiceNo']].nunique().sort_values(by='InvoiceNo', ascending=False).head(5)
```

```python
'''
2.) ลูกค้าจากประเทศใดมีการใช้จ่ายสูงที่สุด (คำนวนจากรายจ่าย)
hint: มีการสร้าง column ใหม่ จากการคูณคอลัมน์อื่น
'''

df['Total'] = df['UnitPrice'] * df['Quantity']
df.groupby('Country')[['Total']].sum().sort_values(by='Total', ascending=False).index[0]
```

```python
'''
3.) จงสรุปยอดซื้อในแต่ละชั่วโมง (คำนวนจากรายจ่าย)
hint: pd.to_datetime
'''

df['InvDate'] = pd.to_datetime(df['InvoiceDate'])
df['hour'] = df['InvDate'].dt.hour
df.groupby('hour')[['Total']].sum()
```

```python
'''
4.) จงหา top 5 สินค้าใดที่มีจำนวนการขายมากที่สุดในเดือนธันวาคม
'''
df['month'] = df['InvDate'].dt.month
df[df['month'] == 12].groupby('Description')[['Quantity']].sum().sort_values(by='Quantity', ascending=False).head(5)
```

```python
'''
5.) สินค้าที่มีคำว่า CHRISTMAS ขายได้ดีที่สุดในเดือนใด (คำนวณจากจำนวนสินค้า)
hint: df['xxx'].str.contains
'''
df[df['Description'].str.contains('CHRISTMAS')].groupby('month')[['Quantity']].sum().sort_values(by='Quantity', ascending=False).head(1)
```

```python
'''
6.) สินค้าที่มีคำว่า CHRISTMAS กับ WOODEN สินค้าใดขายดีสุดในเดือนธันวาคม
'''

temp = df[((df['Description'].str.contains('CHRISTMAS')) & (df['Description'].str.contains('WOODEN'))) & (df['month'] == 12)]
temp.groupby('Description')[['Quantity']].sum().sort_values(by='Quantity', ascending=False).head(1)
```
